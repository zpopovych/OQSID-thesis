## Non-Markovian system identification вЂ” full Kossakowski form, separate Hamiltonians

We fit the model:
$$\dot\rho(t) = \mathcal{L}_{\mathrm{Kos}}(H, C)\,\rho(t) + \mathcal{L}_{\mathrm{Kos}}(H_m, C_m)\,\rho(t - k\Delta t)$$

where the Kossakowski super-operator is
$$\mathcal{L}_{\mathrm{Kos}}(H, C)\,\rho = -i[H,\rho] + \frac{1}{2}\sum_{i,j=1}^{3} C_{ij}\bigl(2 f_i \rho f_j^\dagger - \rho f_j^\dagger f_i - f_j^\dagger f_i \rho\bigr)$$

with $\{f_1, f_2, f_3\} = \{\sigma_x/2, \sigma_y/2, \sigma_z/2\}$ the standard Kossakowski ONB.

The current-time and memory super-operators have **independent Hamiltonians** $H$ and $H_m$
and independent Kossakowski matrices $C$ and $C_m$, each constrained to be positive semidefinite.

Free variables:
- $H$: $\epsilon,\, h_{\mathrm{Re}},\, h_{\mathrm{Im}}$ (3 parameters)
- $H_m$: $\epsilon_m,\, h_{m\mathrm{Re}},\, h_{m\mathrm{Im}}$ (3 parameters)
- $C$: $\gamma_1, \gamma_2, \gamma_3, a_1, a_2, a_3$ (6 parameters)
- $C_m$: $\gamma_{m1}, \gamma_{m2}, \gamma_{m3}, a_{m1}, a_{m2}, a_{m3}$ (6 parameters)

**Total: 18 parameters.**

**Constraints** (Sylvester criterion for $C \succeq 0$ and $C_m \succeq 0$): the three principal minors of each Kossakowski matrix must be non-negative.

```julia
# Kossakowski orthonormal basis: f_i = Пѓ_i / 2
# satisfies tr(fбµў fв±ј) = Оґбµўв±ј/2 and tr(fбµў) = 0
ПѓЛЈ = [ 0 1
       1 0 ]
ПѓКё = [ 0.   im
      -im   0. ]
Пѓб¶» = [ 1.  0
       0  -1. ]

fбґ·в‚Ѓ = ПѓЛЈ / 2
fбґ·в‚‚ = ПѓКё / 2
fбґ·в‚ѓ = Пѓб¶» / 2

@assert tr(fбґ·в‚Ѓ * fбґ·в‚‚) в‰€ 0  &&  tr(fбґ·в‚Ѓ * fбґ·в‚ѓ) в‰€ 0  &&  tr(fбґ·в‚‚ * fбґ·в‚ѓ) в‰€ 0
@assert tr(fбґ·в‚Ѓ * fбґ·в‚Ѓ) в‰€ 1/2  &&  tr(fбґ·в‚‚ * fбґ·в‚‚) в‰€ 1/2  &&  tr(fбґ·в‚ѓ * fбґ·в‚ѓ) в‰€ 1/2

fбґ·бґјбґєбґ® = [fбґ·в‚Ѓ, fбґ·в‚‚, fбґ·в‚ѓ]
```

```julia
"""Kossakowski RHS (no control): -i[H,ПЃ] + (1/2) ОЈбµўв±јCбµўв±ј(2fбµўПЃfв±јвЂ  - ПЃfв±јвЂ fбµў - fв±јвЂ fбµўПЃ)"""
function kossak_rhs(ПЃ, H, C, Fбґјбґєбґ®)
    U = (H * ПЃ - ПЃ * H) / im
    D = sum(C .* [2*fбµў*ПЃ*fв±ј' - ПЃ*fв±ј'*fбµў - fв±ј'*fбµў*ПЃ for fбµў in Fбґјбґєбґ®, fв±ј in Fбґјбґєбґ®]) / 2
    return U + D
end
```

```julia
"""Non-Markovian Kossakowski objective using Simpson quadrature.
Current-time term uses H, C; memory term (k steps back) uses independent Hm, Cm."""
function non_mark_kossak_obj(ПЃ, t, H, C, Hm, Cm, k)
    obj = 0
    for i in (k+3):length(ПЃ)
        ПЃ_avg = (ПЃ[i-2]   + 4ПЃ[i-1]   + ПЃ[i])   / 3
        ПЃ_mem = (ПЃ[i-k-2] + 4ПЃ[i-k-1] + ПЃ[i-k]) / 3
        residual = ПЃ[i] - ПЃ[i-2] -
                   (t[i]-t[i-1]) * kossak_rhs(ПЃ_avg, H,  C,  fбґ·бґјбґєбґ®) -
                   (t[i]-t[i-1]) * kossak_rhs(ПЃ_mem, Hm, Cm, fбґ·бґјбґєбґ®)
        obj += LiPoSID.frobenius_norm2(residual)
    end
    if isempty(monomials(obj))
        obj = 0.
    else
        obj = sum(real(coef) * mon for (coef, mon) in zip(coefficients(obj), monomials(obj)))
    end
    return obj
end
```

```julia
"""Simulate non-Markovian Kossakowski dynamics with RK4.
Memory uses simulated history k steps back; for i в‰¤ k the initial state ПЃ[1] is used."""
function simulate_nonmark_kossak_rk4(ПЃв‚Ђ, t, H, C, Hm, Cm, k)
    ПЃ = Vector{Matrix{ComplexF64}}(undef, length(t))
    ПЃ[1] = ПЃв‚Ђ
    for i in 2:length(t)
        О”t    = t[i] - t[i-1]
        i_mem = max(1, i - k)
        ПЃ_mem = ПЃ[i_mem]
        Lm    = kossak_rhs(ПЃ_mem, Hm, Cm, fбґ·бґјбґєбґ®)
        f(ПЃ_c) = kossak_rhs(ПЃ_c, H, C, fбґ·бґјбґєбґ®) + Lm
        k1 = f(ПЃ[i-1])
        k2 = f(ПЃ[i-1] + О”t/2 * k1)
        k3 = f(ПЃ[i-1] + О”t/2 * k2)
        k4 = f(ПЃ[i-1] + О”t     * k3)
        ПЃ[i] = ПЃ[i-1] + О”t/6 * (k1 + 2k2 + 2k3 + k4)
    end
    return ПЃ
end
```

```julia
"""Constrained sparse SOS optimisation via TSSOS.
Minimises obj subject to constr[i] в‰Ґ 0 for all i.
Returns (solution, optimal_value, status)."""
function cs_tssos_opt(obj::Polynomial, constr::Vector{<:Polynomial})
    all_vars = union(variables(obj), reduce(union, variables.(constr)))
    solution   = variables(obj) => nothing
    opt_val    = nothing
    status     = :Error
    relax_order = maxdegree(obj) > 2 ? maxdegree(obj) : 2
    try
        redirect_stdout(open("/dev/null", "w")) do
            redirect_stderr(open("/dev/null", "w")) do
                opt, sol, data = TSSOS.cs_tssos_first(
                    [obj; constr...], all_vars, relax_order;
                    numeq=0, solution=true, QUIET=true)
                prev_opt, prev_sol, prev_data = opt, sol, data
                while sol !== nothing
                    prev_opt, prev_sol, prev_data = opt, sol, data
                    opt, sol, data = TSSOS.cs_tssos_higher!(data; QUIET=true, solution=true)
                end
                ref_sol, flag = TSSOS.refine_sol(prev_opt, prev_sol, prev_data;
                                                 QUIET=true, tol=1e-10)
                solution = variables(obj) => ref_sol
                opt_val  = prev_opt
                status   = flag == 0 ? :Global : :Local
            end
        end
    catch e
        println("TSSOS failed: ", e)
    end
    return solution, opt_val, status
end
```

### Symbolic setup

Independent Hamiltonians for current-time and memory terms:
$$H = \frac{1}{2}\begin{pmatrix} \epsilon & h_{\mathrm{Re}} + i h_{\mathrm{Im}} \\ h_{\mathrm{Re}} - i h_{\mathrm{Im}} & -\epsilon \end{pmatrix}, \qquad
H_m = \frac{1}{2}\begin{pmatrix} \epsilon_m & h_{m\mathrm{Re}} + i h_{m\mathrm{Im}} \\ h_{m\mathrm{Re}} - i h_{m\mathrm{Im}} & -\epsilon_m \end{pmatrix}$$

Kossakowski matrix parameterisation (Theorem 3.1, GKS 1976):
$$C = \begin{pmatrix} -\gamma_1+\gamma_2+\gamma_3 & -ia_3 & ia_2 \\ ia_3 & \gamma_1-\gamma_2+\gamma_3 & -ia_1 \\ -ia_2 & ia_1 & \gamma_1+\gamma_2-\gamma_3 \end{pmatrix}$$

$C \succeq 0$ iff the Sylvester minors $\kappa_i \geq 0$,  $\kappa_1\kappa_2+\kappa_2\kappa_3+\kappa_3\kappa_1 \geq a_1^2+a_2^2+a_3^2$, and $\det C \geq 0$, where $\kappa_i$ are the diagonal entries. Same conditions apply to $C_m$.

```julia
# --- Current Hamiltonian ---
@polyvar Пµ h_Re h_Im

H0ЛўКёбµђбµ‡ = [ Пµ                h_Re + im*h_Im
            h_Re - im*h_Im  -Пµ              ] / 2

# --- Memory Hamiltonian ---
@polyvar Пµm hm_Re hm_Im

HmЛўКёбµђбµ‡ = [ Пµm                hm_Re + im*hm_Im
            hm_Re - im*hm_Im  -Пµm              ] / 2

# --- Current Kossakowski matrix C ---
@polyvar Оі[1:3] a[1:3]

Оєв‚Ѓ = -Оі[1]+Оі[2]+Оі[3]
Оєв‚‚ =  Оі[1]-Оі[2]+Оі[3]
Оєв‚ѓ =  Оі[1]+Оі[2]-Оі[3]

CЛўКёбµђбµ‡ = [ Оєв‚Ѓ       -im*a[3]   im*a[2]
           im*a[3]   Оєв‚‚       -im*a[1]
          -im*a[2]   im*a[1]   Оєв‚ѓ      ]

# PSD constraints for C
c_constr1 = Оі[1] + Оі[2] + Оі[3]                                          # tr(C)/2 в‰Ґ 0
c_constr2 = Оєв‚Ѓ*Оєв‚‚ + Оєв‚‚*Оєв‚ѓ + Оєв‚ѓ*Оєв‚Ѓ - a[1]^2 - a[2]^2 - a[3]^2         # sum of 2Г—2 minors в‰Ґ 0
c_constr3 = Оєв‚Ѓ*Оєв‚‚*Оєв‚ѓ - Оєв‚Ѓ*a[1]^2 - Оєв‚‚*a[2]^2 - Оєв‚ѓ*a[3]^2             # det(C) в‰Ґ 0

# --- Memory Kossakowski matrix Cm ---
@polyvar Оіm[1:3] am[1:3]

Оєmв‚Ѓ = -Оіm[1]+Оіm[2]+Оіm[3]
Оєmв‚‚ =  Оіm[1]-Оіm[2]+Оіm[3]
Оєmв‚ѓ =  Оіm[1]+Оіm[2]-Оіm[3]

CmЛўКёбµђбµ‡ = [ Оєmв‚Ѓ        -im*am[3]    im*am[2]
            im*am[3]    Оєmв‚‚        -im*am[1]
           -im*am[2]    im*am[1]    Оєmв‚ѓ      ]

# PSD constraints for Cm
cm_constr1 = Оіm[1] + Оіm[2] + Оіm[3]
cm_constr2 = Оєmв‚Ѓ*Оєmв‚‚ + Оєmв‚‚*Оєmв‚ѓ + Оєmв‚ѓ*Оєmв‚Ѓ - am[1]^2 - am[2]^2 - am[3]^2
cm_constr3 = Оєmв‚Ѓ*Оєmв‚‚*Оєmв‚ѓ - Оєmв‚Ѓ*am[1]^2 - Оєmв‚‚*am[2]^2 - Оєmв‚ѓ*am[3]^2

constraints = [Оі[1], Оі[2], Оі[3], c_constr1, c_constr2, c_constr3,
               Оіm[1], Оіm[2], Оіm[3], cm_constr1, cm_constr2, cm_constr3]
```

```julia
data_dir   = "../DATA/"
models_dir = "../MODELS/"
tests_dir  = "../TESTS/"

dodeca_files = ["State_D"*string(n) for n=1:20]
basis_files  = ["State_B"*string(n) for n=1:4]

train_files = basis_files
test_files  = dodeca_files

k = 5  # memory steps back (П„ = kВ·О”t)
```

```julia
function g_nonmark_kossak_objective(Оібµў)
    obj = 0
    for df_trn in train_files
        ПЃбµ—КівЃї, tбµ—КівЃї = LiPoSID.get_rho_series(data_dir*df_trn*"_2CUT_data.h5", Оібµў)
        end_train   = min(length(tбµ—КівЃї), 1200)
        ПЃбµ—КівЃї = convert(Vector{Matrix{ComplexF64}}, ПЃбµ—КівЃї[1:end_train])
        tбµ—КівЃї = convert(Vector{Float64},            tбµ—КівЃї[1:end_train])
        obj += non_mark_kossak_obj(ПЃбµ—КівЃї, tбµ—КівЃї, H0ЛўКёбµђбµ‡, CЛўКёбµђбµ‡, HmЛўКёбµђбµ‡, CmЛўКёбµђбµ‡, k)
    end
    return obj
end
```

```julia
Оі = ["0.079477", "0.25133", "0.79477", "2.5133", "7.9477", "25.133", "79.477", "251.33"]

date_and_time_string = string(Dates.format(now(), "yyyy-u-dd_at_HH-MM"))
tests_data_file_name = "NonMark_Kossak2H_k$(k)_trn4_tst20_"*date_and_time_string*".h5"

subs_num(sym, sol)  = begin v = subs(sym, sol); try; convert(Float64,    v);      catch; 0.0;      end; end
subs_cmat(mat, sol) = map(mat) do e; v = subs(e, sol); try; convert(ComplexF64, v); catch; 0.0+0im; end; end

"""Project matrix to nearest valid density matrix: Hermitian, PSD, trace-1."""
function to_density_op(basis, ПЃ)
    ПЃh = Hermitian((ПЃ + ПЃ') / 2)
    F  = eigen(ПЃh)
    vals = max.(real(F.values), 0.0)
    s = sum(vals)
    s > 0 && (vals ./= s)
    DenseOperator(basis, Matrix(Hermitian(F.vectors * Diagonal(vals) * F.vectors')))
end

for Оібµў in Оі
    println("Оі = "*Оібµў)

    poly = g_nonmark_kossak_objective(Оібµў)
    sol, opt_val, status = cs_tssos_opt(poly, constraints)
    println("  status: ", status, "  opt_val: ", opt_val)

    HЛўвЃ±бµ€  = subs_cmat(H0ЛўКёбµђбµ‡, sol)
    HmЛўвЃ±бµ€ = subs_cmat(HmЛўКёбµђбµ‡, sol)
    CЛўвЃ±бµ€  = subs_cmat(CЛўКёбµђбµ‡,  sol)
    CmЛўвЃ±бµ€ = subs_cmat(CmЛўКёбµђбµ‡, sol)

    epsilon   = subs_num(Пµ,  sol)
    epsilon_m = subs_num(Пµm, sol)
    println("  Оµ=$(epsilon)  Оµm=$(epsilon_m)")

    h5open(tests_dir*tests_data_file_name, "cw") do fid
        Оі_group = create_group(fid, "gamma_"*Оібµў)
        Оі_group["H"]         = HЛўвЃ±бµ€
        Оі_group["Hm"]        = HmЛўвЃ±бµ€
        Оі_group["C"]         = CЛўвЃ±бµ€
        Оі_group["Cm"]        = CmЛўвЃ±бµ€
        Оі_group["k"]         = k
        Оі_group["epsilon"]   = epsilon
        Оі_group["epsilon_m"] = epsilon_m
    end

    for df in test_files
        print(df*" ")

        ПЃв‚›, tв‚› = LiPoSID.get_rho_series(data_dir*df*"_2CUT_data.h5", Оібµў)
        ПЃв‚› = convert(Vector{Matrix{ComplexF64}}, ПЃв‚›)
        tв‚› = convert(Vector{Float64}, tв‚›)

        ПЃЛўвЃ±бµ€ = simulate_nonmark_kossak_rk4(ПЃв‚›[1], tв‚›, HЛўвЃ±бµ€, CЛўвЃ±бµ€, HmЛўвЃ±бµ€, CmЛўвЃ±бµ€, k)

        ПЃбµ—Лўбµ—    = [DenseOperator(basis, Hermitian(ПЃв‚њ))  for ПЃв‚њ in ПЃв‚›]
        ПЃЛўвЃ±бµ€_op = [to_density_op(basis, ПЃв‚њ)             for ПЃв‚њ in ПЃЛўвЃ±бµ€]

        Fв‚‘в‚“ = [abs(fidelity(ПЃв‚Ѓ, ПЃв‚‚)) for (ПЃв‚Ѓ, ПЃв‚‚) in zip(ПЃбµ—Лўбµ—, ПЃЛўвЃ±бµ€_op)]

        h5open(tests_dir*tests_data_file_name, "cw") do fid
            Оі_group          = open_group(fid, "gamma_"*Оібµў)
            init_state_group = create_group(Оі_group, df)
            init_state_group["Fidelity"] = convert.(Float64, Fв‚‘в‚“)
        end

        println(minimum(Fв‚‘в‚“))
    end

    println()
end

println(tests_data_file_name)
```
