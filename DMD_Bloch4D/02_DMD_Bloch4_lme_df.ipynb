```python
from boson_data_lib import *
import numpy as np
import scipy.linalg as sl
import os
import pandas as pd
import glob
```

```python
best_init_rho =  np.array([[ 0.0+0.j,  0.0+0.0j],
                            [ 0.0-0.0j, 1.0+0.0j] ]) # exited state

data_dir = "../DATA/"
fid_dists = fidelity_distances(data_dir, best_init_rho);
```

```python
def read_as_dataframe(filename, tests_dir, fid_dists):

    df = pd.DataFrame()
    
    print("Processing...")
    
    with h5py.File(tests_dir+filename, "r") as f:          
        for gamma in f.keys():
            
            print(gamma)
            
            g = gamma[6:]

            states = ["State_D"+str(i+1) for i in range(20)]

            for state in states:

                F_b4dmd = f[gamma][state]["F_b4dmd_lme"][()]

                print(min(F_b4dmd))

                F_b4a = f[gamma][state]["F_b4a_lme"][()] 

                t = f[gamma][state]["time"][()]
                ser_len = len(t)

                gamma_column = [g] * ser_len
                state_column = [state] * ser_len

                time_column = t
                gamma_time_column = np.array(t)*float(g) 

                fid_dist_column = [fid_dists[int(state[7:])-1]] * ser_len 

                d_ser = {'Gamma': gamma_column,
                        'State': state_column,
                        'Distance': fid_dist_column,
                        'Time': time_column,
                        'gt': gamma_time_column,
                        'Fidelity_b4DMD': F_b4dmd[:ser_len],
                        'Infidelity_b4DMD': 1-F_b4dmd[:ser_len],
                        'Fidelity_b4a': F_b4a[:ser_len],
                        'Infidelity_b4a': 1-F_b4a[:ser_len]}

                df_ser = pd.DataFrame(data = d_ser)   
                df = pd.concat([df, df_ser])
    
    print(" done!")
    
    return df
    
```

```python
tests_dir = "../TESTS/"
tests_file = "DMD-Bloch4D_LME_trn4_tst20_2023-Oct-07_at_13-50.h5"

tests_dir
```

```python
df = read_as_dataframe(tests_file, tests_dir, fid_dists)
```
