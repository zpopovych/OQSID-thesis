```python
import h5py
```

```python
tests_dir = "../TESTS/"
#test_file = "POP_LME_6_random_trn4_tst20_2024-Jan-16_at_11-29.h5" #simple Hamiltonian
#test_file = "POP_LME_6_random_trn4_tst20_2024-Jan-17_at_11-32.h5" #simple Hamiltonian
#test_file = "POP_LME_5_random_trn4_tst20_2024-Jan-16_at_16-36.h5"
test_file = "POP_LME_5_random_trn4_tst20_2024-Jan-20_at_15-06.h5"
```

```python
def read_gammas(filename, tests_dir):
    
    with h5py.File(tests_dir+filename, "r") as f:
    
        gammas = ['0.079477', '0.25133', '0.79477', '2.5133', '7.9477', '25.133', '79.477', '251.33']
        
        relaxation = []
        de_phasing = []
        #exitation = []
        de_polarisation = []
        omega = []

               
        for gamma in gammas:

            relaxation.append(f["gamma_"+gamma]["gamma_relaxation"][()])
            de_phasing.append(f["gamma_"+gamma]["gamma_de_phasing"][()])
            #exitation.append(f["gamma_"+gamma]["gamma_exitation"][()])
            de_polarisation.append(f["gamma_"+gamma]["gamma_de_polarisation"][()])
            omega.append(f["gamma_"+gamma]["omega"][()])
            
            print(gamma)

    return gammas, omega, relaxation, de_phasing, de_polarisation  # exitation
    
```

```python
gammas, omega, relaxation, de_phasing, de_polarisation = read_gammas(test_file, tests_dir)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(omega)
```

```python
plt.plot(de_phasing)
```

```python


plt.semilogy(gammas,omega/max(omega),  label="omega")
plt.semilogy(gammas,relaxation/max(relaxation),  label="relaxation")
plt.semilogy(de_polarisation/max(de_polarisation), label="depolarization")
plt.semilogy(de_phasing/max(de_phasing), label="dephasing")
#plt.semilogy(exitation/max(exitation), label="exitation")
plt.legend()
```

```python
plt.semilogy(gammas,relaxation,  label=r"$\gamma_1$ (relaxation)")
plt.semilogy(de_phasing, label=r"$\gamma_2$ (dephasing)")
#plt.semilogy(exitation, label=r"$\gamma_3$ (excitation)")
plt.semilogy(de_polarisation, label=r"$\gamma_4$ (depolarization)")
plt.xlabel(r'Coupling, $\gamma$')
#plt.title('Markovian processes in spin-boson system')
plt.legend()
plt.savefig("Processes_5_in_spin-boson.pdf")
```
