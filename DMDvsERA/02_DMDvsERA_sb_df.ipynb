```python
from boson_data_lib import *
import numpy as np
import scipy.linalg as sl
import os
import pandas as pd
import glob
```

```python
def read_as_dataframe(filename, tests_dir):

    df = pd.DataFrame()
    
    print("Processing...")
    
    with h5py.File(tests_dir+filename, "r") as f:

        for rank in f.keys():

            print("Rank = ", rank[5:])
        
            #gammas = ['0.0','0.079477', '0.25133', '0.79477', '2.5133', '7.9477', '25.133', '79.477', '251.33']
            
            for gamma in f[rank].keys():
                
                print(gamma)
                
                g = gamma[6:]

                states = ["State_D"+str(i+1) for i in range(20)]

                for state in states:

                    F_dmd = f[rank][gamma][state]["F_dmd_sb"][()]

                    print(min(F_dmd))

                    F_era = f[rank][gamma][state]["F_era_sb"][()] 

                    t = f[rank][gamma][state]["time"][()]
                    ser_len = len(t)

                    rank_column  = [rank[5:]] * ser_len
                    gamma_column = [g] * ser_len
                    state_column = [state] * ser_len

                    d_ser = {'Rank': rank_column,
                            'Gamma': gamma_column,
                            'State': state_column,
                            'Fidelity_DMD': F_dmd[:ser_len],
                            'Fidelity_ERA': F_era[:ser_len]}

                    df_ser = pd.DataFrame(data = d_ser)   
                    df = pd.concat([df, df_ser])
    
    print(" done!")
    
    return df
```

```python
tests_dir = "../TESTS/"
tests_file = "DMDvsERA_SB_trn4_tst20_2023-Sep-19_at_11-07.h5"
df = read_as_dataframe(tests_file, tests_dir)
```
