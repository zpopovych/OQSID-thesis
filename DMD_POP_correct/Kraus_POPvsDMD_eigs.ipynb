```python
import h5py
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d_lme =[]
d_sb = []

obj_val_DMD_sb = []
obj_val_POP_sb = []

with h5py.File("MODELS/POP-Kraus_vs_DMD_trn4_rm_micro_obj_models_2023-Sep-16_at_18-37.h5", "r") as f:

    for gamma in f.keys():

        obj_val_DMD_sb.append(f[gamma]["obj_val_DMD_sb"][()])
        obj_val_POP_sb.append(f[gamma]["obj_val_POP_sb"][()])

```

```python
d_lme =[]
d_sb = []

obj_val_DMD_sb = []
obj_val_POP_sb = []

with h5py.File("MODELS/POP-Kraus_vs_DMD_trn4_models_2023-Sep-13_at_14-05.h5", "r") as f:

    for gamma in f.keys():

        d_lme.append(sum(f[gamma]["d_QPT_LME"][()] < 0))
        d_sb.append(sum(f[gamma]["d_QPT_sb"][()] < 0))

        obj_val_DMD_sb.append(f[gamma]["obj_val_DMD_sb"][()])
        obj_val_POP_sb.append(f[gamma]["obj_val_POP_sb"][()])

df = pd.DataFrame({
    'Gamma': ['0.079', '0.251', '0.794', '2.513', '7.947', '25.13', '79.47', '251.3'],
    'LME': d_lme,
    'spin-boson': d_sb
})

fig, ax1 = plt.subplots()

df.plot(ax=ax1, x="Gamma", xlabel=r"Coupling, $\gamma$", y=["LME", "spin-boson"], kind="bar", yticks=[0,1,2], ylabel=r"Number of negative eigenvalues of $\chi$ matrix")
fig.tight_layout()
fig.savefig("QPT-Kraus_of_DMD_choi_eigs.pdf")
```

```python
plt.plot(obj_val_POP_sb, label="POP")
plt.plot(obj_val_DMD_sb, label="DMD")
plt.xticks(range(8), ['0.079', '0.251', '0.794', '2.513', '7.947', '25.133', '79.477', '251.33'])
plt.legend()
plt.title("Value of objective function with identified Kraus operators")
```
