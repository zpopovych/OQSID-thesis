```python
import h5py
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
def read_as_dataframe(filename, tests_dir):

    df = pd.DataFrame()
    
    print("Processing...")
    
    with h5py.File(tests_dir+filename, "r") as f:

        for gamma in f.keys():

            print("\n", gamma)
        
            #gammas = ['0.0','0.079477', '0.25133', '0.79477', '2.5133', '7.9477', '25.133', '79.477', '251.33']
            
            #F_POP_vs_sb = [] 
            #F_POP_vs_LME = [] 
            #F_DMD_vs_sb = [] 
            #F_DMD_vs_LME = [] 
            #F_DMD_Kraus_vs_sb = [] 
            #F_DMD_Kraus_vs_LME = [] 

            for state in f[gamma].keys():
                
                print(state, end =" ")
                
                g = gamma[6:]

                F_POP_vs_sb = f[gamma][state]["F_POP-vs-sb"][()]
                F_POP_vs_LME = f[gamma][state]["F_POP-vs-LME"][()]
                F_DMD_vs_sb = f[gamma][state]["F_DMD-vs-sb"][()]
                F_DMD_vs_LME = f[gamma][state]["F_DMD-vs-LME"][()]
                F_DMD_Kraus_vs_sb = f[gamma][state]["F_DMD-Kraus-vs-sb"][()]
                F_DMD_Kraus_vs_LME = f[gamma][state]["F_DMD-Kraus-vs-LME"][()]

                #t = f[gamma][state]["time"][()]

                ser_len = len(F_POP_vs_sb)

                #print(ser_len)

                g = gamma[6:]
               
                gamma_column = [g] * ser_len
                state_column = [state] * ser_len
                #time_column = t[:ser_len]
                #gamma_time_column = np.array(t[:ser_len])*float(g) 


                d_ser = {'Gamma': gamma_column,
                        'State': state_column,
                        'F_POP_vs_sb' : F_POP_vs_sb,
                        'F_POP_vs_LME':F_POP_vs_LME,
                        'F_DMD_vs_sb' :F_DMD_vs_sb,
                        'F_DMD_vs_LME' :F_DMD_vs_LME,
                        'F_DMD_Kraus_vs_sb' : F_DMD_Kraus_vs_sb, 
                        'F_DMD_Kraus_vs_LME' : F_DMD_Kraus_vs_LME
                        }
                
                        #'Time': time_column, 
                        #'gt': gamma_time_column,

                df_ser = pd.DataFrame(data = d_ser)   
                df = pd.concat([df, df_ser])
    
    print(" done!")
    
    return df
```

```python
parent_dir = "/home/zah/github/OQSID-thesis/"
data_dir = parent_dir + "DATA/"
tests_dir = parent_dir + "TESTS/"

tests_file = "POP-Kraus_vs_DMD_trn4-tst20_obj_fids_2023-Dec-20_at_13-56.h5"
df = read_as_dataframe(tests_file, tests_dir)
```
